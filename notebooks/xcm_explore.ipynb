```python
import matplotlib.pyplot as plt
import soundfile as sf
import numpy as np


ax = plt.figure(figsize=(20,20)).add_subplot(projection='3d')
for i in range(0,20):
    data, sr = sf.read(dataset['train'][i]['filepath'])
    ax.plot(np.arange(0,len(data)),data,zs=i,zdir='y')
```

```python
from datasets import load_dataset
import numpy as np
import librosa
import librosa.display
import soundfile as sf
import matplotlib.pyplot as plt 
from IPython.display import Audio
```

```python
# load dataset 
dataset = load_dataset('DBD-research-group/BirdSet', 'HSN', trust_remote_code=True)
```

```python
dataset['train'] 
```

```python
data, samplerate = sf.read(dataset['train'][0]['filepath'])
```

```python

# display first n samples 
for i in range(10):
    # some plotting 
    data, samplerate = sf.read(dataset['train'][i]['filepath'])
    plt.plot(data)
    Audio(data, rate=samplerate)
    
    print(dataset['train'].features)
    # computing / test
    D = np.abs(librosa.stft(data))**2
    S = librosa.feature.melspectrogram(S=D, sr=samplerate)
    
    # display the calculated spectrograms for train purposes 
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(librosa.power_to_db(S, ref=np.max), sr=samplerate, y_axis='mel', x_axis='time')
    plt.title('Mel Spectrogram')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
    
    duration = len(data) / samplerate
    
    # not an "analysis" at this point, just some examination of data
    mean_amplitude = np.mean(data)
    variance_amplitude = np.var(data)
    
    print(f'Duration: {duration:.2f} seconds')
    print(f'Mean Amplitude: {mean_amplitude:.4f}')
```
